```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
import string

def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def preprocess_text(text):
    # Remove punctuation and convert to lowercase
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator).lower()
    words = cleaned_text.split()
    return words

def calculate_statistics(words):
    total_words = len(words)
    unique_words = len(set(words))
    average_word_length = sum(len(word) for word in words) / total_words
    word_counts = Counter(words)
    return total_words, unique_words, average_word_length, word_counts

def visualize_data(word_counts):
    # Most frequent words
    freq_words = word_counts.most_common(10)
    words, counts = zip(*freq_words)

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.barh(words, counts, color='lightblue', edgecolor='black')  # Horizontal bars with specified colors
    plt.xlabel('Frequency', fontsize=12)  # Increased font size for labels
    plt.ylabel('Words', fontsize=12)
    plt.title('Top 10 Most Frequent Words', fontsize=14)  # Increased font size for title
    plt.show()


    df = pd.DataFrame(freq_words, columns=['Word', 'Count'])
    print("Most Frequent Words Table:")
    print(df)

def main():
    file_path = '/content/LiberationWar_BD.txt'
    text = read_file(file_path)
    words = preprocess_text(text)
    total_words, unique_words, avg_word_length, word_counts = calculate_statistics(words)
    print(f"Total Words: {total_words}")
    print(f"Unique Words: {unique_words}")
    print(f"Average Word Length: {avg_word_length:.2f}")
    visualize_data(word_counts)

if __name__ == "__main__":
    main()
```
